```python
# 12-Stu_Challenge_Activity

# Increase = Current Price - Original Price
# Percent Increase = Increase / Original x 100

# Create variable for original_price

original_price = 360.36


# Create variable for current_price

current_price = 500.00

# Create float for threshold_to_buy

threshold_to_buy = -20.00


# Create float for threshold_to_sell

threshold_to_sell = 20.00

# Create float for portfolio balance

portfolio_balance = 10000

# Create float for balance check

balance_check = portfolio_balance * 5


# Create string for recommendation, default will be buy

recommendation = "unavailable"

# Calculate difference between current_price and original_price

price_change = current_price - original_price


# Calculate percent change

percent_change = (current_price/original_price) * 100

# Print original_price

print(f"Netflix's original stock price was ${original_price}")

# Print current_price

print(f"Netflix's current stock price is ${current_price}")


# Print percent increase

print(f"Netflix's stock price changed by {percent_change:.2f}%")

# Determine if stock should be bought or sold

if percent_change >= threshold_to_sell:
    recommendation = "Sell"
elif percent_change <= threshold_to_buy:
    recommendation = "Buy"
elif percent_change < threshold_to_buy and percent_change > threshold_to_sell:
    recommendation = "Hold"
else: 
    print("Not enough data is available to make a decision. Human Input is Required")

# Print recommendation
print()
print("Recommendation: " + recommendation)
print()

# Balance Check

print(f"You currently have ${portfolio_balance} of liquid capital available in your portfolio.")
if recommendation == "Buy" and balance_check > current_price:
    print("Final Recommendation: " + recommendation )
    print()
elif recommendation == "Buy" and balance_check < current_price:
    print("Final Recommendation is to buy. However, you do not have the available capital")
else:  
    print()
    print("Final Recommendation: " + recommendation)
    


```
